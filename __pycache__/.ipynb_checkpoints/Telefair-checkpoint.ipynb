```python
import telebot
from ser import iter_ivent
tok = '1342963300:AAGBnoqmIQ63KpcVMa9QuPgF_RtF_HXhwEU'
bot = telebot.TeleBot(tok, parse_mode=None)

types = telebot.types
```

```python
idd = ''
hideBoard = types.ReplyKeyboardRemove() # 

def come_in():
    if come_in == True:
        return 'Online'
    elif come_in == False:
        return 'Offline'
@bot.message_handler(commands=['start', 'back'])
def start(message):
    markup = types.ReplyKeyboardMarkup(row_width=2)
#     markup1 = types.ReplyKeyboardMarkup(row_width=1)
    itembtn1 = types.KeyboardButton('Login')
    itembtn2 = types.KeyboardButton('Inplay ({0})'.format(come_in()))
    itembtn3 = types.KeyboardButton('Soon ({0})'.format(come_in()))
    itembtn4 = types.KeyboardButton('Exit')
    markup.add(itembtn1, itembtn2, itembtn3, itembtn4)
#     markup1.add(itembtn1, itembtn4)
    bot.send_message(message.from_user.id,'Ok', reply_markup=markup)
#     bot.send_message(message.from_user.id,'Ok', reply_markup=markup1)

@bot.message_handler(func=lambda message: True)
def echo_all(message):    
    if message.text == "Login":
        markup = types.ReplyKeyboardMarkup(row_width=1)
        itembtn1 = types.KeyboardButton(message.from_user.id)
        itembtn2 = types.KeyboardButton('/back')
        markup.add(itembtn1, itembtn2)
        bot.send_message(message.from_user.id, "Id chat:", reply_markup=markup)
    elif message.text == 'Inplay ({0})'.format(come_in()):
        markup = types.ReplyKeyboardMarkup(row_width=1)
        itembtn1 = types.KeyboardButton('Inplay')
        itembtn2 = types.KeyboardButton('/back')
        markup.add(itembtn1, itembtn2)
        bot.send_message(message.from_user.id, text=next(iter_ivent) ,reply_markup=markup)
    elif message.text == 'Soon ({0})'.format(come_in()):
#         inl =types.InlineKeyboardButton('Go', url='https://mail.zoho.com/zm/#mail/folder/inbox/p/1611831160247100001')
        markup = types.ReplyKeyboardMarkup(row_width=1)
        itembtn1 = types.KeyboardButton('Soon')
        itembtn2 = types.KeyboardButton('/back')
        markup.add(itembtn1, itembtn2)
        bot.send_message(message.from_user.id, 'Ok', reply_markup=markup)
    elif message.text == 'Exit':
        bot.send_message(message.from_user.id, "Пока", reply_markup=hideBoard)
        bot.stop_polling()
        
        
# #         bot.send_message(message.from_user.id, "Привет, чем я могу тебе помочь?")
#     elif message.text == "off" :
#         bot.send_message(message.from_user.id, "Пока", reply_markup=hideBoard)
#         bot.stop_polling()
#     else:
#         bot.send_message(message.from_user.id, "Я тебя не понимаю. Напиши /start.")
# #     bot.reply_to(message, message.text)

bot.polling(none_stop=False, interval=0)
```
